### **Discrete-Event Simulation (DES) using SimPy**

#### **Define a Process Function/Generator**

```python
def car(env):
    while True:
        print(f'Car starts parking at {env.now}')
        parking_duration = 5
        yield env.timeout(parking_duration)

        print(f'Car starts driving at {env.now}')
        driving_duration = 2
        yield env.timeout(driving_duration)
```

#### **Creating Model Instance and Observing Behavior**

```python
import simpy

env = simpy.Environment()       # creates an environment object
env.process(car(env))           # starts the car process

env.run(until=30)
```

### **Learnings #1**
1. Creating a model is as simple as three steps: 1) creating environment, 2) define process/es, and 3) runnning the model (assuming there are pre-defined generators already).
